```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv('/Users/praveenkumardevamane/Downloads/Dynamic pricing/Dynamic-Pricing/data/sales_data.csv')
data.info()

data.shape
```

```python
df=data.drop(['Product ID','Store ID'],axis=1).copy()
df.info()
```

```python
df.describe()

```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
print(df['Date'])
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Date', y='Demand', data=df, label='Demand')
sns.lineplot(x='Date', y='Price', data=df, label='Price')
plt.title("Demand vs Price Over Time")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Inventory Level', y='Demand', data=df)
plt.title("Demand vs Inventory Level")
plt.xlabel("Inventory Level (Supply)")
plt.ylabel("Demand")
plt.show()
```
